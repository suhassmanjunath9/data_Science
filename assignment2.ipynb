```python
import random

def estimate_conditional_probability(num_trials):
    # Define the contents of the bag.
    bag = ['red'] * 5 + ['green'] * 7 + ['blue'] * 8
    
    # Initialize counters for the simulation.
    blue_count = 0
    red_count = 0
    blue_followed_by_red_count = 0
    red_followed_by_blue_count = 0

    # Simulate the process of drawing a ball 1000 times.
    previous_draw = None
    for _ in range(num_trials):
        current_draw = random.choice(bag)
        
        # Count the occurrences of each color.
        if current_draw == 'blue':
            blue_count += 1
        if current_draw == 'red':
            red_count += 1
            
        # Check for conditional occurrences.
        if previous_draw == 'blue' and current_draw == 'red':
            blue_followed_by_red_count += 1
        if previous_draw == 'red' and current_draw == 'blue':
            red_followed_by_blue_count += 1
            
        previous_draw = current_draw
    # Estimate P(Blue) from simulation
    p_blue_sim = blue_count / num_trials
    
    # Estimate P(Red) from simulation
    p_red_sim = red_count / num_trials

    p_red_given_blue_sim = blue_followed_by_red_count / (blue_count - (1 if previous_draw == 'blue' else 0))
    
    p_blue_given_red_sim = red_followed_by_blue_count / (red_count - (1 if previous_draw == 'red' else 0))

    return p_red_given_blue_sim, p_blue_sim, p_red_sim, p_blue_given_red_sim

# --- Main execution ---
number_of_trials = 100000
p_red_given_blue_sim, p_blue_sim, p_red_sim, p_blue_given_red_sim = estimate_conditional_probability(number_of_trials)

# Theoretical probabilities for comparison
total_balls = 5 + 7 + 8
p_red_theory = 5 / total_balls
p_blue_theory = 8 / total_balls
p_red_given_blue_theory = p_red_theory # Since the events are independent
p_blue_given_red_theory = p_blue_theory # Since the events are independent

print(f"Simulation based on {number_of_trials} trials:")
print("-" * 40)
print(f"Estimated P(Red | Blue) from simulation: {p_red_given_blue_sim:.4f}")
print(f"Estimated P(Blue | Red) from simulation: {p_blue_given_red_sim:.4f}")
print(f"Estimated P(Red) from simulation: {p_red_sim:.4f}")
print(f"Estimated P(Blue) from simulation: {p_blue_sim:.4f}")

# Verify Bayes' Theorem with simulation results
# P(Red|Blue) â‰ˆ [P(Blue|Red) * P(Red)] / P(Blue)
# Since the events are independent, P(Red|Blue) = P(Red). Let's see if the simulation holds.
bayes_verification_sim = (p_blue_given_red_sim * p_red_sim) / p_blue_sim
print("\nVerifying Bayes' Theorem with simulation data:")
print(f"Simulated P(Red|Blue): {p_red_given_blue_sim:.4f}")
print(f"Simulated [P(Blue|Red) * P(Red)] / P(Blue): {bayes_verification_sim:.4f}")

print("\nTheoretical probabilities for comparison:")
print("-" * 40)
print(f"Theoretical P(Red|Blue): {p_red_given_blue_theory:.4f}")
print(f"Theoretical P(Blue|Red): {p_blue_given_red_theory:.4f}")
print(f"Theoretical P(Red): {p_red_theory:.4f}")
print(f"Theoretical P(Blue): {p_blue_theory:.4f}")
```

```python
import numpy as np

def analyze_discrete_random_variable(sample_size):
    
    # Define the possible outcomes (values) and their corresponding probabilities.
    outcomes = [1, 2, 3]
    probabilities = [0.25, 0.35, 0.4]

    sample = np.random.choice(outcomes, size=sample_size, p=probabilities)
    empirical_mean = np.mean(sample)
    empirical_variance = np.var(sample)
    empirical_std_dev = np.std(sample)
    
    return empirical_mean, empirical_variance, empirical_std_dev

# Define the size of the sample.
sample_size = 1000

# Run the simulation and get the results.
mean, variance, std_dev = analyze_discrete_random_variable(sample_size)

# Print the results to the console.
print(f"Sample size: {sample_size}")
print(f"Empirical Mean: {mean:.4f}")
print(f"Empirical Variance: {variance:.4f}")
print(f"Empirical Standard Deviation: {std_dev:.4f}")
```

```python
import numpy as np

import matplotlib.pyplot as plt



def visualize_exponential_distribution(sample_size, mean):

    sample = np.random.exponential(scale=mean, size=sample_size)

    

    # Calculate empirical mean and standard deviation for comparison.

    empirical_mean = np.mean(sample)

    empirical_std_dev = np.std(sample)

    

    print(f"Sample size: {sample_size}")

    print(f"Mean of theoretical distribution: {mean}")

    print(f"Empirical mean of the sample: {empirical_mean:.4f}")

    print(f"Empirical standard deviation of the sample: {empirical_std_dev:.4f}")

    

    # --- Visualization ---

    

    # Create a figure and an axes object.

    fig, ax = plt.subplots(figsize=(10, 6))

    ax.hist(sample, bins=50, density=True, alpha=0.6, color='skyblue', label='Histogram of Sample Data')

    

    # Overlay the theoretical PDF.

    # Generate x-values for the PDF curve.

    x = np.linspace(0, np.max(sample), 100)

    

    # The PDF of an exponential distribution is $f(x; \lambda) = \lambda e^{-\lambda x}$.

    # Since $\lambda = 1/\text{mean}$, the formula becomes $f(x; \text{mean}) = (1/\text{mean})e^{-x/\text{mean}}$.

    pdf = (1 / mean) * np.exp(-x / mean)

    

    # Plot the PDF curve.

    ax.plot(x, pdf, 'r-', linewidth=2, label=f'Theoretical PDF (mean={mean})')



    # Add labels, title, and legend.

    ax.set_title(f'Histogram and PDF of an Exponential Distribution (mean={mean})')

    ax.set_xlabel('Value')

    ax.set_ylabel('Probability Density')

    ax.grid(True, linestyle='--', alpha=0.6)

    ax.legend()

    

    # Display the plot.

    plt.show()



# Set the parameters for the simulation.

sample_size = 2000

distribution_mean = 5



# Run the visualization function.

visualize_exponential_distribution(sample_size, distribution_mean)

```

```python
import numpy as np
import matplotlib.pyplot as plt

def central_limit_theorem_simulation(population_size, num_samples, sample_size):
    """
    Simulates the Central Limit Theorem (CLT) by drawing samples from a uniform
    distribution and visualizing the distribution of the sample means.

    The Central Limit Theorem states that the distribution of sample means
    will approximate a normal distribution, regardless of the shape of the
    population distribution, as the sample size grows.

    Args:
        population_size (int): The number of data points in the initial uniform distribution.
        num_samples (int): The number of samples to draw from the population.
        sample_size (int): The size of each sample.
    """
    # Step a: Generate a large population of random numbers from a uniform distribution.
    # We use a range of 0 to 1 for simplicity.
    population = np.random.uniform(low=0.0, high=1.0, size=population_size)

    # Step b: Draw multiple samples and calculate the mean of each sample.
    sample_means = []
    for _ in range(num_samples):
        # Draw a random sample of a specified size from the population.
        sample = np.random.choice(population, size=sample_size)
        # Calculate the mean of the current sample and add it to our list.
        sample_means.append(np.mean(sample))

    # Convert the list of means to a NumPy array for easier calculations.
    sample_means = np.array(sample_means)

    # Print some statistics for the sample means to the console.
    print(f"Population size: {population_size}")
    print(f"Number of samples: {num_samples}")
    print(f"Sample size (n): {sample_size}")
    print("-" * 40)
    print(f"Mean of sample means: {np.mean(sample_means):.4f}")
    print(f"Standard deviation of sample means: {np.std(sample_means):.4f}")

    # Step c: Visualize the distribution of the sample means.
    # Create a figure with two subplots side by side.
    fig, axes = plt.subplots(1, 2, figsize=(15, 6))

    # Plot 1: Histogram of the original uniform population distribution.
    axes[0].hist(population, bins=50, density=True, color='lightgreen', edgecolor='black', alpha=0.7)
    axes[0].set_title('Original Uniform Distribution')
    axes[0].set_xlabel('Value')
    axes[0].set_ylabel('Probability Density')
    axes[0].grid(True, linestyle='--', alpha=0.6)

    # Plot 2: Histogram of the sample means distribution.
    # This will show the bell-shaped curve predicted by the CLT.
    axes[1].hist(sample_means, bins=50, density=True, color='skyblue', edgecolor='black', alpha=0.7)
    axes[1].set_title('Distribution of Sample Means')
    axes[1].set_xlabel('Sample Mean')
    axes[1].set_ylabel('Probability Density')
    axes[1].grid(True, linestyle='--', alpha=0.6)

    # Add a main title for the entire figure.
    fig.suptitle('Central Limit Theorem Simulation')

    # Ensure a tight layout to prevent labels from overlapping.
    plt.tight_layout(rect=[0, 0.03, 1, 0.95])
    
    # Display the plots.
    plt.show()

# Set the parameters for the simulation.
population_size = 10000
num_samples = 1000
sample_size = 30  # A sample size of 30 or more is often considered sufficient for CLT.

# Run the simulation function.
central_limit_theorem_simulation(population_size, num_samples, sample_size)
```
