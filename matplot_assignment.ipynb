```python
import matplotlib.pyplot as plt
import numpy as np


#1: Create a simple line plot
# Define the data for the line plot.
x = [1, 2, 3, 4, 5]
y = [10, 15, 25, 30, 50]

# a. Plot the data.
plt.figure(figsize=(8, 6)) # Optional: Set the figure size
plt.plot(x, y, marker='o', linestyle='-', color='b')

# b. Customize the plot by adding a title, axis labels, and a grid.
plt.title('Simple Line Plot', fontsize=16)
plt.xlabel('X-axis', fontsize=12)
plt.ylabel('Y-axis', fontsize=12)
plt.grid(True, which='both', linestyle='--', linewidth=0.5)

# Display the plot.
plt.show()



# 2: Create a bar graph

# Define the data for the bar graph.
students = ['John', 'Jane', 'Alice', 'Bob']
marks = [75, 85, 60, 90]

# a. Plot the data as a bar graph.
plt.figure(figsize=(8, 6))
colors = ['skyblue', 'salmon', 'lightgreen', 'gold']
plt.bar(students, marks, color=colors)

# b. Customize the colors and add a title.
plt.title('Student Marks in a Subject', fontsize=16)
plt.xlabel('Students', fontsize=12)
plt.ylabel('Marks', fontsize=12)

# Display the plot.
plt.show()


# 3: Create a pie chart
# Define the data for the pie chart.
regions = ['North America', 'Europe', 'Asia', 'Others']
revenue = [45, 25, 20, 10]

# a. Create a pie chart with the region names as labels.
plt.figure(figsize=(8, 8))

# b. Highlight the region with the highest revenue.
explode = [0] * len(revenue)
highest_revenue_index = revenue.index(max(revenue))
explode[highest_revenue_index] = 0.1

# Create the pie chart with the labels, percentages, and explode effect.
plt.pie(revenue, labels=regions, autopct='%1.1f%%', startangle=90, explode=explode,
        colors=['#66c2a5', '#fc8d62', '#8da0cb', '#e78ac3'],
        shadow=True)

plt.title('Company Revenue Distribution by Region', fontsize=16)

# Display the plot.
plt.show()

#4: Generate a histogram
# Generate 1000 random integers between 1 and 100.
data = np.random.randint(1, 101, 1000)

# Create the histogram.
# `bins` determines the number of intervals for the frequency distribution.
plt.figure(figsize=(10, 6))
plt.hist(data, bins=10, edgecolor='black', alpha=0.7)

# Add title and labels.
plt.title('Frequency Distribution of Random Integers', fontsize=16)
plt.xlabel('Value', fontsize=12)
plt.ylabel('Frequency', fontsize=12)

# Display the plot.
plt.show()
```
